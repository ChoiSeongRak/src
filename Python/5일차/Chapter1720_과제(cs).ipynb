[문제6]

ZeroDivisionError,  IndexError 외에 3가지 이상의 파이썬 에러에 대해 설명해주세요

1. TypeError : 잘못된 데이터 타입이 사용되었을 때
2. NameError : 정의되지 않은 변수나 함수를 사용할 때
3. KeyError : 딕셔너리에 존재하지 않는 키를 사용하려 할 때
