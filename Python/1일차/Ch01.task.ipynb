<span style="font-size:80%">[문제6]<br><br></span>
```Python에서 메모리 관리는 어떻게 이루어지는지 설명해주세요```

<span style="color:red">답변</span>

파이썬 인터프리터를 통해 자동으로 이루어진다.

그 중 참조 카운트란 객체의 참조 카운트를 사용하는데 객체가 생성될 때마다

해당 객체에 대한 참조 카운트가 1 증가하며 다른 객체가 이를 참조할 때마다

해당 객체의 카운트가 1 증가, 사라질 때마다 카운트 감소

카운트가 0이 되면 객체가 메모리에서 해제됨

이 외에도

-순환 참조를 감지하여 메모리 누수 방지
-가비지 컬렉터로 더 이상 참조되지 않는 객체를 자동으로 해제
