[문제1]

list와 tuple의 주된 차이점 은 무엇인가요?

1. 가변성<br>
    list: 가변 데이터로 요소를 추가, 삭제, 변경 가능<br>
    tuple: 불변 데이터로 한번 생성되면 요소 변경 불가능

2. 구문<br>
    list: 대괄호 [] 를 사용하여 정의<br>
    tuple: 소괄호 () 를 사용하여 정의
